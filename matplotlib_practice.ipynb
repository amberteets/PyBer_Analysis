```python
# Import dependencies.
# Chaining .pyplot to matplotlib is same as 'from matplotlib import pyplot'
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot.
plt.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
    # Unpack a tuple containing a figure and axex object(s) to the variables fig, ax
        # ax allows us to annotate the chart
    # Then plot data from x- and y- axes 
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt(), alternate method
# Used to change figure-level attributes (such as axis labels, a title, or a legend) or save the figure as an image
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot and use plt.show() method.
#plt.plot(x_axis, y_axis)
#plt.show()
```

```python
# Create the plot and add a label for the legend.
plt.plot(x_axis, y_axis, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
# Create the plot.
plt.plot(x_axis, y_axis, marker="d", color="blue", linewidth=1.5, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add a grid.
plt.grid()
# Add the legend.
plt.legend()
```

```python
# Create the plot and make below modifications using object-oriented interface method
fig = plt.figure()
ax = fig.add_subplot()
# Change the line color to green.
# Change the marker to a diamond.
# Change the line width to 2 points.
# Add a legend for the city of Boston.
ax.plot(x_axis, y_axis, label="Boston", marker="D", color="green", linewidth=2)
# Add a title and axes labels.
ax.set_xlabel('Date')
ax.set_ylabel('Fare ($)')
ax.set_title("PyBer Fare by Month")
# Add a y-axis limit.
ax.set_ylim(0, 45)
# Add grid lines.
ax.grid()
```

Bar Charts

```python
plt.bar(x_axis, y_axis)
```

```python
# Create the plot.
plt.bar(x_axis, y_axis, color="green", label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
plt.barh(x_axis, y_axis, color="magenta", label="Boston")
plt.xlabel("Fare ($)")
plt.ylabel("Date")
plt.title("PyBer Fare by Month")
plt.legend()
# Use .gca() to 'get current axes' then invert the y-axis so Jan comes first
plt.gca().invert_yaxis()
```

```python
# Create plot (bar chart) using object oriented interface (OOI) method
fig, ax = plt.subplots()
ax.bar(x_axis, y_axis)
```

```python
# Create horizontal bar chart using OOI method
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis, color="teal", label="Chicago")
ax.legend()
ax.set_xlabel("Fare ($)")
ax.set_ylabel("Date")
ax.set_title("PyBer Fare by Month")
plt.gca().invert_yaxis()
```

SCATTER PLOTS

```python
# Create scatter plot using MATLAB .plot() method
plt.plot(x_axis, y_axis, 'o')
```

```python
# Create scatter plot using MATLAB .scatter() method
plt.scatter(y_axis, x_axis, color="red", label="Chicago")
plt.title("PyBer Fare by Month")
plt.xlabel("Fare ($)")
plt.ylabel("Date")
plt.xlim(0, 50)
plt.legend()
plt.gca().invert_yaxis()
```

```python
# Bubble Charts
plt.scatter(x_axis, y_axis, s = y_axis)
```

```python
# To make markers bigger, scale up the source data (without actually changing the source data)
y_axis_scaled = []
for data in y_axis:
    y_axis_scaled.append(data*3)

# Set size parameter to scaled up axis
plt.scatter(x_axis, y_axis, s = y_axis_scaled)
```

```python
# Use list comprehension within .scatter() method to scale the y_axis values and plot the bubble chart.
plt.scatter(x_axis, y_axis, s = [y*3 for y in y_axis])
```

```python
# OOI Method for Scatter Plots
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis)
```

```python
# OOI Method for Bubble Plots
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis, s=y_axis)
```

```python
# OOI Method for Bubble Plots
fig, ax = plt.subplots()
# Change the color of the markers to sky blue.
# Change the size of the markers to 5 times each data point.
# Make the color 20% transparent.
# Add a black edge color to the circles.
# Make the linewidth of the circles 2 points.
# Add a legend for the city of Boston.
# Switch the axis so that the Fare($) data is on the x-axis.
ax.scatter(y_axis, x_axis, s=[5*i for i in y_axis], color="deepskyblue", edgecolors="black", alpha=.8, linewidth=2, label="Boston")

# Add a title.
ax.set_title("PyBer Fare by Month")
ax.set_xlabel("Fare ($)")
ax.set_ylabel("Date")
# Add a limit for the x-axis data.
ax.set_xlim(0, 50)
# Invert the y-axis so that January is at the top.
plt.gca().invert_yaxis()
```

```python
# Pie Chart using MATLAB Method
plt.pie(y_axis, labels=x_axis)
#plt.show()
```

```python
# Increase figure size (must be first line of code)
plt.subplots(figsize=(8, 8))
# Add 'explode' parameter and percentages for each month.
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.pie(y_axis, explode=explode_values, labels=x_axis, autopct='%.1f%%')
```

```python
# Assign 12 colors, one for each month.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8, 8))
plt.pie(y_axis,
        explode=explode_values,
        colors=colors,
        labels=x_axis,
        autopct="%.1f%%")

plt.show()
```

```python
# Object-Oriented Interface Approach for Pie Charts
explode_values = (0, 0, .2, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
fig, ax = plt.subplots(figsize=(8,8))
ax.pie(y_axis,labels=x_axis,
      autopct='%.1f%%',
      explode=explode_values,
      shadow=True,
      startangle=90,
      counterclock=False,
      colors=colors)

plt.show()
```
